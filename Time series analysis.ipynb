# Time Series Analysis

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import mplfinance as mpf
from datetime import date, timedelta
```

```python
# Define date ranges
today = date.today()
end_date = today.strftime("%Y-%m-%d")
start_date = (today - timedelta(days=720)).strftime("%Y-%m-%d")
```

```python
# Download stock data
data = yf.download('AAPL', start=start_date, end=end_date, progress=False)
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'])
plt.title('Time Series Analysis (Line Plot)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
# Candlestick Chart using mplfinance
mpf.plot(data, type='candle', title='Time Series Analysis (Candlestick Chart)', ylabel='Price')
```

```python
# Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(data.index, data['Close'])
plt.title('Time Series Analysis (Bar Plot)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
# Custom Date Range Line Plot
custom_data = data.loc['2021-07-01':'2021-12-31']
plt.figure(figsize=(10, 6))
plt.plot(custom_data.index, custom_data['Close'])
plt.xticks(rotation=90)
plt.title('Time Series Analysis (Custom Date Range)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```
